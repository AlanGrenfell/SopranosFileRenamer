Problem: I have The Sopranos saved on my computer, but the episode titles are missing and the titles are not very readable, like this:

        "The.Sopranos.S03E05.1080p.5.1Ch.BluRay.ReEnc-DeeJayAhmed.mkv"

I want it to include the episode titles, which often bear significance to the content and as I am a big fan, I identify episodes with their titles. I'd also like it to be more readable. 

Solution: In this notebook I scrape wikipedia for the episode titles, then reformat the file names to include them.

```python
import os
import urllib.request
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_The_Sopranos_episodes"
page = urllib.request.urlopen(url)

soup = BeautifulSoup(page, "lxml")

#print(soup.prettify()) 
#<table class="wikitable plainrowheaders wikiepisodetable" style="width:100%">
#<th id="ep1" rowspan="1" scope="row" style="text-align:center">

all_tables = soup.find_all("table")
#print(all_tables)

right_table = soup.find_all("table", class_ = "wikitable plainrowheaders wikiepisodetable")
#right_table
```

```python
  
df1 = pd.read_html(str(right_table))[0]
df2 = pd.read_html(str(right_table))[1]
df3 = pd.read_html(str(right_table))[2]
df4 = pd.read_html(str(right_table))[3]
df5 = pd.read_html(str(right_table))[4]
df6 = pd.read_html(str(right_table))[5]

df = pd.concat([df1,df2,df3, df4, df5, df6 ], ignore_index=True)
df
```

```python
episode_list = df['Title'].tolist()

for item in range(len(episode_list)):
    episode_list[item] = episode_list[item].replace("\\", "")
    episode_list[item] = episode_list[item].replace('\"', "")
    episode_list[item] = episode_list[item].strip('\/:*?"<>|') #stripping illegal characters for filenames in win10

    
episode_list.remove("Part 1") #correcting minor error of web scraping
episode_list.remove("Part 2") #where season 6 parts were taken as titles

    
index_list = [i for i in range(len(episode_list))]

ep_dict = {index_list[i]: episode_list[i] for i in range(len(index_list))} 

#ep_dict        
```

```python
path = 'G:\Media\Video\Sopranos'
os.chdir(path)

count = 0

for f in os.listdir():
    path = 'G:\Media\Video\Sopranos'
    path = path + "\\" + f
    os.chdir(path)
    for g in os.listdir():
        f_name, f_ext = (os.path.splitext(g))
        if f_ext == ".srt":
            continue
        count += 1
        f_tit1, f_tit2, f_epno, f_res, f_5, f_garb, f_blu, f_idiot = (f_name.split('.'))
        f_title = ep_dict[(count-1)]
        dash = "-"
        new_name = '{} {} {} {} {}{}'.format(f_epno, f_title, dash, f_tit1, f_tit2, f_ext)
        print(new_name)
        try:
            os.rename(g, new_name)
        except:
            print(g)
```
